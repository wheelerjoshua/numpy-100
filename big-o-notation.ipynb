```python
#import pyplot module as alias plt
import matplotlib.pyplot as plt  

#import other libraries 
import numpy as np
import math
from random import randint
import random
import pandas as pd
```

```python
#Curves to graph
n = [n for n in range(1,5)]
y1 = [0*n + 1 for n in n]
y2 = [math.log(n) for n in n]
y3 = [n for n in n]
y4 = [n * math.log(n) for n in n]
y5 = [n**2 for n in n]
y6 = [2 ** n for n in n]
y7 = [math.factorial(n) for n in n]
y8 = [n ** n for n in n]
```

```python
# Establish figure
plt.figure(figsize = (10,10))

#Individual plots for each curve, with color and label
plt.plot(n, y1, color = 'b', label = 'O(1)')
plt.plot(n, y2, color = 'r', label = 'O(log n)')
plt.plot(n, y3, color = 'g', label = 'O(n)')
plt.plot(n, y4, color = 'orange', label = 'O(n log n)')
plt.plot(n, y5, color = 'k', label = '$O(n^2)$')
plt.plot(n, y6, color = 'deepskyblue', label = '$O(2^n)$')
plt.plot(n, y7, color = 'y', label = 'O(n!)')
plt.plot(n, y8, color = 'magenta', label = '$O(n^n)$')


# X and Y axis labels
plt.xlabel('Elements', color = 'w')
plt.ylabel('Operations', color = 'w')
# Change tick color to be visible on dark background
plt.xticks(color = 'white')
plt.yticks(color = 'white')

# Plot legend
plt.legend();
```
