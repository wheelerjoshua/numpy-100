### Imports

```python
#import pyplot module as alias plt
import matplotlib.pyplot as plt  

#import other libraries 
import numpy as np
import math
from random import randint
import random
import pandas as pd
```

### 1.

```python
# Create list for x from 0 to 100
x = [x for x in range(0,100)]

# Create variable y with ist comprehension to create a list of values where x is run 
# through a function
y = [x**2 - x +2 for x in x]

# Establish figure for plotting
plt.figure(figsize = (10,10))

# Create plot of x and y with color black
plt.plot(x,y, color = 'k')

# Annotate origin point
plt.annotate('This point is the origin, 0,0', xy = (0,0), xytext = (0,5000), color = 'r', fontsize = 14, arrowprops = dict(facecolor='b'))
```

### 2.

```python
# Establish variables for x and various functions
x = [x for x in range(0,5)]
y1 = [math.sqrt(x) for x in x]
y2 = [x**3 for x in x]
y3 = [2**x for x in x]
y4 = [1/(x+1) for x in x]

# Plot individual functions in separate figures
plt.figure(figsize = (5,5))
plt.plot(x,y1)
plt.xlabel('x')
plt.ylabel('y = sqrt(x)')


plt.figure(figsize = (5,5))
plt.plot(x,y2)
plt.xlabel('x')
plt.ylabel('y = x^3')


plt.figure(figsize = (5,5))
plt.plot(x,y3)
plt.xlabel('x')
plt.ylabel('y = 2^x')

plt.figure(figsize = (5,5))
plt.plot(x,y4)
plt.xlabel('x')
plt.ylabel('y = 1/(x + 1)')
```

### 3.

```python
# Establish figure
plt.figure(figsize = (10,10))


# Plot separate axes for each function, but in the same figure, using subplots
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.xlabel('x')
plt.ylabel('y = sqrt(x)')


plt.subplot(2,2,2)
plt.plot(x,y2)
plt.xlabel('x')
plt.ylabel('y = x^3')


plt.subplot(2,2,3)
plt.plot(x,y3)
plt.xlabel('x')
plt.ylabel('y = 2^x')


plt.subplot(2,2,4)
plt.plot(x,y4)
plt.xlabel('x')
plt.ylabel('y = 1/(x + 1)')

# Add space for each subplot
plt.tight_layout()
```

### 4.

```python
# Establish figure
plt.figure(figsize=(8,8))

# Plotting functions on the same axes
plt.plot(x,y1, color = 'b',
        label = '$\sqrt{x}$')


plt.plot(x,y2, color = 'r',
        label = '$x^3$')


plt.plot(x,y3, color = 'g', 
        label = '$2^x$')


plt.plot(x,y4, color = 'orange', 
         label = '1 / (x + 1)')
#        label = "$\frac{1}{(x + 1)$")
# $\frac$ causes an error

# Plot legend
plt.legend();
```
