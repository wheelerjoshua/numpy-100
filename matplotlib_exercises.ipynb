### Imports

```python
#import pyplot module as alias plt
import matplotlib.pyplot as plt  

#import other libraries 
import numpy as np
import math
from random import randint
import random
import pandas as pd
```

### 1.

```python
x = [x for x in range(0,100)]
y = [x**2 - x +2 for x in x]
plt.figure(figsize = (10,10))
plt.plot(x,y, color = 'k')
plt.annotate('This point is the origin, 0,0', xy = (0,0), xytext = (0,5000), color = 'r', fontsize = 14, arrowprops = dict(facecolor='b'))
```

### 2.

```python
x = [x for x in range(0,10)]
y1 = [math.sqrt(x) for x in x]
y2 = [x**3 for x in x]
y3 = [2**x for x in x]
y4 = [1/(x+1) for x in x]

plt.figure(figsize = (5,5))
plt.plot(x,y1)
plt.xlabel('x')
plt.ylabel('y = sqrt(x)')


plt.figure(figsize = (5,5))
plt.plot(x,y2)
plt.xlabel('x')
plt.ylabel('y = x^3')


plt.figure(figsize = (5,5))
plt.plot(x,y3)
plt.xlabel('x')
plt.ylabel('y = 2^x')

plt.figure(figsize = (5,5))
plt.plot(x,y4)
plt.xlabel('x')
plt.ylabel('y = 1/(x + 1)')
```
