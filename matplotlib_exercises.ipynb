### Imports

```python
#import pyplot module as alias plt
import matplotlib.pyplot as plt  

#import other libraries 
import numpy as np
import math
from random import randint
import random
import pandas as pd
```

### 1.

```python
# Create list for x from -100 to 100
x = [x for x in range(-100,100)]

# Create variable y with ist comprehension to create a list of values where x is run 
# through a function
y = [x**2 - x +2 for x in x]

# Establish figure for plotting
plt.figure(figsize = (10,10))

# Create plot of x and y with color black
plt.plot(x,y, color = 'k')

# Annotate origin point
plt.annotate('This point is the origin, 0,0', xy = (0,0), 
             xytext = (0,5000), 
             color = 'r', 
             fontsize = 14, 
             arrowprops = dict(facecolor='b'))
```

### 2.

```python
# Establish variables for x and various functions
x1 = range(0,100)
y1 = [math.sqrt(x) for x in x1]

# Plot individual functions in separate figures

plt.figure(figsize = (5,5))
plt.title('$y = \sqrt{x}$')
plt.plot(x1,y1)
plt.xlabel('x')
plt.ylabel('$\sqrt{x}$')

x2 = [x for x in range(0,20)]
y2 = [x**3 for x in x2]
plt.figure(figsize = (5,5))
plt.title('$y = x^3$')
plt.plot(x2,y2)
plt.xlabel('x')
plt.ylabel('$x^3$')

x3 = [x * 0.1 for x in range(0,150)]
y3 = [2**x for x in x3]
plt.figure(figsize = (5,5))
plt.title('$y = 2^x$')
plt.plot(x3,y3)
plt.xlabel('x')
plt.ylabel('$2^x$')

x4 = [x * 0.1 for x in range(0,150)]
y4 = [1/(x+1) for x in x4]
plt.figure(figsize = (5,5))
plt.title('y = 1/(x + 1)')
plt.plot(x4,y4)
plt.xlabel('x')
plt.ylabel('1 / (x + 1)')
```

### 3.

```python
# Establish figure
plt.figure(figsize = (10,10))


# Plot separate axes for each function, but in the same figure, using subplots
plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.xlabel('x')
plt.ylabel('y = sqrt(x)')


plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.xlabel('x')
plt.ylabel('y = x^3')


plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.xlabel('x')
plt.ylabel('y = 2^x')


plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.xlabel('x')
plt.ylabel('y = 1/(x + 1)')

# Change tick color to be visible on dark background
plt.xticks(color = 'white')
plt.yticks(color = 'white')

# Add space for each subplot
plt.tight_layout()
```

### 4.

```python
# Establish figure
plt.figure(figsize=(8,8))

# Plotting functions on the same axes
plt.plot(x1,y1, color = 'b',
        label = '$\sqrt{x}$')


plt.plot(x2,y2, color = 'r',
        label = '$x^3$')


plt.plot(x3,y3, color = 'g', 
        label = '$2^x$')


plt.plot(x4,y4, color = 'orange', 
         label = '1 / (x + 1)')
#        label = "$\frac{1}{(x + 1)$")
# $\frac$ causes an error

# adjust xlim and ylim
plt.ylim(0, 10)
plt.xlim(0, 10)


# Plot legend
plt.legend();
```
